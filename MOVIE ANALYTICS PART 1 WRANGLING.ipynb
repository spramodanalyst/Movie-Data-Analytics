The first stage in any EDA problem solving strategy is Data cleaning  and Manipulating stage.

Data Cleaning and Manipulation Steps:

1. **Data Understanding:**
   Reading the dataset documentation Importing data - Understanding each coloumn data Data summarisation like checking the number of rows and coloumns of the dataset, checking the datatype of each coloumn etc.

2. **Data Cleaning**:
   Dropping the irrelevant coloumns, Renaming the coloumns, Dropping the duplicate rows, Dropping or handling missing values, Dropping invalid data rows, Detecting and handling outliers.

3. **Data Manipulation:**
   Coloumn transformation, Joining datasets, other manipulation like pivoting or transposing.
